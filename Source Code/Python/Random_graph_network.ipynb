```python
import networkx as nx
import matplotlib.pyplot as nzm
%matplotlib inline
```

```python
import os
os.listdir()
```

```python
#read in file
p = nx.read_edgelist('com-amazon.top5000.cmty.txt',create_using=nx.Graph(),nodetype=int)
```

```python
print(nx.info(p))
```

```python
#create a layout of fb
pos = nx.spring_layout(p)
nzm.figure(figsize=(15,12))
nzm.axis('on')
nx.draw_networkx(p,pos=pos,with_labels=False,node_size=50)
```
