```python
import collections
import networkx as nx
import matplotlib.pyplot as nzm
%matplotlib inline
```

```python
import os
os.listdir()
```

```python
#read in file
p = nx.read_edgelist('politician_edges_txt.txt',create_using=nx.Graph(),nodetype=int)
```

```python
print(nx.info(p))
```

```python
#create a layout of fb
pos = nx.spring_layout(p)
nzm.figure(figsize=(15,12))
nzm.axis('on')
nx.draw_networkx(p,pos=pos,with_labels=False,node_size=50)
```

```python
#draw a bar chart of the betwenness centrality
bc = nx.betweenness_centrality(p)
nzm.bar(range(len(bc)),bc.values(),align='center')
```

```python
nx.betweenness_centrality(p, k=None, normalized=True, weight=None, endpoints=False, seed=None)
```

```python
G = nx.gnp_random_graph(100, 0.02)

degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = nzm.subplots()
nzm.bar(deg, cnt, width=0.80, color='b')

nzm.title("Degree Histogram")
nzm.ylabel("Count")
nzm.xlabel("Degree")
ax.set_xticks([d + 0.4 for d in deg])
ax.set_xticklabels(deg)

# draw graph in inset
nzm.axes([0.4, 0.4, 0.5, 0.5])
Gcc = sorted(nx.connected_component_subgraphs(G), key=len, reverse=True)[0]
pos = nx.spring_layout(G)
nzm.axis('off')
nx.draw_networkx_nodes(G, pos, node_size=20)
nx.draw_networkx_edges(G, pos, alpha=0.4)

nzm.show()
```
